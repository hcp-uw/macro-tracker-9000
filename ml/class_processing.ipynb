```python
import polars as pl
```

```python
with open('raw_classes.txt', 'r') as f:
    contents = f.read()
```

```python
pages = contents.split('\n\n')
```

```python
class_nums = []
names = []
nums = []
for page in pages:
    for line in page.split('\n')[1:]:
        for i, word in enumerate(line.split()[1:]):
            try:
                int(word.replace(',', ''))
                break
            except:
                pass
    
        row_1 = line.split()[:i + 2]
        row_2 = line.split()[i + 2:]
        print(row_1, row_2)
        class_num, name, num = int(row_1[0]), ' '.join(row_1[1:-1]), int(row_1[-1].replace(',', ''))
        class_nums.append(class_num)
        names.append(name)
        nums.append(num)
        class_num, name, num = int(row_2[0]), ' '.join(row_2[1:-1]), int(row_2[-1].replace(',', ''))
        class_nums.append(class_num)
        names.append(name)
        nums.append(num)
```

```python
pl.DataFrame({"class": class_nums, "name": names, "num_instances": nums}, schema={"class": pl.Int64, "name": pl.String, "num_instances": pl.Int64}).write_csv("classes.csv")
```
